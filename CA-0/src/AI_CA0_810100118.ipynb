<h5 style="color:red;">DataFrame.fillna(value=None, *, method=None, axis=None, inplace=False, limit=None, downcast=None)</h5>


<pre style="color:blue;">To fill NaN values , some variables have<b style="color:red;"> discrete</b> value and some of them have<b style="color:red;"> continuis</b> value , for example sex has only two value like MALE and FEMALE (that with label encoding they transform to 1 or 0) so to fill NaN value of sex you cannot put mean in them like 1.4 :))))<br><b> Deleting the data is definitely a very bad option </b>


- advantages :
    It reduces the variance
    No significant loss of data (~0.1%)
    Easy to implement
    Preserves the distribution
    Filling NaN values with the mean does not introduce any bias in the data
    

- disadvantages :
    Can reduce variance
    Can distort correlation
    <b>** Not suitable for categorical data</b>
    For a particular cateory, all null values present for a column
    
    
    

In summary, filling NaN values with the mean is a simple and quick method of imputing missing data that has some advantages, but also some disadvantages. The suitability of this method depends on the nature of the data and the requirements of the subsequent analysis.
    

</pre>

<h1 style="color:green;">Question 8 : Loop vs Vectorize</h1>

```python
def meanWithLoop(df):
    count = 0 
    sumOfHours = 0
    for inx in df.index:
        if (df['education'][inx] == 'Bachelors'):
            sumOfHours += df['hours-per-week'][inx]
            count += 1 
    return (sumOfHours / count)
```

```python
print ("The value with loop iterator is same : {:.5f}" 
      .format(meanWithLoop(trainDF)))
```

<h3 style="color:red;">Now I comapere the Compute Times </h3>

```python
import scipy.stats as stats
```

```python
def drawGraphs (moreThan50Mean , moreThan50Std , lessThan50Mean , lessThan50Std , DF_nor):
    colNum = 0 
    for col in DF_nor:
        moreThan50DistValues = stats.norm(moreThan50Mean[colNum] , moreThan50Std[colNum])
        lessThan50DistValues = stats.norm(lessThan50Mean[colNum] , lessThan50Std[colNum])
        xValue = DF_nor[col].unique()
        xValue.sort()

        plt.plot(xValue, moreThan50DistValues.pdf(xValue), label='More than 50')
        plt.plot(xValue, lessThan50DistValues.pdf(xValue), label='Less than 50')
        plt.legend()
        plt.xlabel(col)
        plt.tick_params(which="minor" , direction = 'in' , length = 4 , width = 1 , labelsize = 10)


        plt.grid(which='major' , axis='both' , linestyle = '--' , color="black")
        plt.grid(which='minor' , axis='both' , linestyle = ':' , color="gray")
        plt.minorticks_on()

        plt.ylabel(f"PDF of {col}")
        plt.title(f"PDF of {col} per {col}")
        plt.tight_layout()
        plt.show()
        colNum +=1
        if (col == 'native-country'):
            break
    
```

```python
drawGraphs(moreThan50Mean , moreThan50Std , lessThan50Mean , lessThan50Std , DF_nor)
```

```python
listOfPred = []
```

```python
def predictionOn (prop , DF_nor , intersectionValue , fromLeftBlueIsUp , trainDF , listOfPred):
    truePred = 0 
    DF_nor[f'pred_on_{prop}'] = np.nan
    if (fromLeftBlueIsUp == 1):    
        DF_nor.loc[(DF_nor[prop] >= intersectionValue) , f'pred_on_{prop}' ] = ">50K"
        DF_nor.loc[(DF_nor[prop] < intersectionValue), f'pred_on_{prop}' ] = "<=50K"
    else :
        DF_nor.loc[(DF_nor[prop] >= intersectionValue) , f'pred_on_{prop}' ]  = "<=50K"
        DF_nor.loc[(DF_nor[prop] < intersectionValue) , f'pred_on_{prop}' ]  = ">50K"
    
    truePred = (DF_nor[ f'pred_on_{prop}'] ==  trainDF['salary']).sum()
         
    #print (f"{prop} : " ,truePred / len(DF_nor.index) * 100)
    listOfPred.append([f"{prop} : " ,truePred / len(DF_nor.index) * 100])
    
```

```python
predictionOn('age',DF_nor ,-1.2,1 , trainDF , listOfPred)

predictionOn('workclass',DF_nor ,3 ,1 , trainDF , listOfPred)

predictionOn('fnlwgt',DF_nor ,0 ,0 , trainDF , listOfPred)

predictionOn('education',DF_nor ,10.5 ,1 , trainDF , listOfPred)

predictionOn('education-num',DF_nor ,0.2 ,1 , trainDF , listOfPred)

predictionOn('marital-status',DF_nor ,2.41 ,0 , trainDF , listOfPred)

predictionOn('occupation',DF_nor ,5.75 ,1 , trainDF , listOfPred)

predictionOn('relationship',DF_nor ,1.2 , 0 , trainDF , listOfPred)

predictionOn('race',DF_nor ,3.7 ,1 , trainDF, listOfPred)

predictionOn('capital-loss',DF_nor ,0.25 ,1 , trainDF, listOfPred)

predictionOn('hours-per-week',DF_nor ,0.15 ,1 , trainDF, listOfPred)

predictionOn('native-country',DF_nor ,36 ,1 , trainDF, listOfPred)

predictionOn('capital-gain',DF_nor ,1.25 ,1 , trainDF, listOfPred)
```

 - The variance of the normal distribution represents the spread of the distribution. If the variance of a property is significantly larger or smaller than the variance of the distribution, it may also be a good predictor of the outcome you are interested in.
